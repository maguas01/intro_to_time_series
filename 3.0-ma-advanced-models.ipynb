## more advanced models 
- capture patterns 
- arima model; box jenkins model; univariate and multivariate; seasonal and non-seasonal
- exponential smothing; captures seasonality

### box jenkins
- has 3 parameter [AR] p(autoregressive) order, [I]d(integration/differencing) degree, and [MA] q(moving average) order
- requires stationary time series
- arima() acf, pacf, and auto.arima() 

p - summation of lags -ar, d - degree of differenceing -I, q - summation of forecast - MA

```r
# arima 

plot(lynx)
```

```r
# autoregression 
acf(lynx); pacf(lynx)
```

```r
# there are several options for models
# find the smallest number of orders (parameter)
# smallest information criteria (AIC, AICc, BIC)
# - zero means or non-zero mean 
library(forecast)
auto.arima(lynx)
```

![arima](img/arima_results.png)

```r
#?auto.arima()
auto.arima(lynx, trace= T)
```

```r
myar = auto.arima(lynx, stepwise=0, approximation = 0)
myar
```

```r
plot(forecast(myar, h=10))
```

## parameters of exponential smothing
- error(additive or multiplicative always present), trend(non-present, additive, multiplicative), seasonality(non-present, addive or multiplicative)
- put more weight into most recent results 
- function ets() selects the best optimal model 

```r
library(forecast)

etsmodel = ets(nottem); etsmodel
# ETS(error, trend, seasonality), A=additive, M=multiplicative, N
# smothing coefficient
# the closer the coefficients are to one the more the model relies on recent data
# alpha - error, gama-seasonality, beta trend
```

```r
# plot the nottem model on top of the forecast 
plot(nottem, lwd=3)
lines(etsmodel$fitted, col="red")
```

```r
plot(forecast(etsmodel, h=36))
```
