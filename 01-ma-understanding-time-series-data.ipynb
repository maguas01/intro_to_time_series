### Lag: the gap between two or more observations 
- difference of some value now - the value of some time ago for example, the lynx data record is 114 as we can see the the function lenght(lynx). $Y_{t}$ is some observation in time series. So the last observation in this case would be $Y_{114} = 3396$. If we want to calculate lag of 1 then $= Y_{t} - Y_{t-1} = Y_{114} - Y_{113} = 3396 - 2657$

### testing for stationarity, does the data have the same statistical properties throughout the time series ? 
- contant mean, variance and no autocorrelation 
- if data is non stationary use a transformation, or differencing (adjust data according to timespan)
- de-trend if trend exist 
### differencing: 
    first difference goes from $Y_{t}$ to the next ie. $Y_{t} - Y_{t-1}$ 
    
    if the 1st difference is stationary and completely random and not autocorrelated then we call this a random walk each value is a random step away from the previous value
    
    if the first difference is stationary but not completely random then we a more advance model such as exponential smoothing 

```r
# stationary 
test = ts(c(rnorm(100,2,1), rnorm(100, 50,1)), start =1)
plot(test)
```

```r
# clearly there is a break the first part has a
# mean of about 2 while the second has a mean of about 50
# clearly we have a non constant mean, and non stationary data
# here differencing is a great option for making our data stationary 
# while logrithmic tranformion will equalize the variance

plot(diff(test))
# we can see from the plot the mean is station
```

```r
# unit root test

# unit root test check if a dataset is stationary and if differencing is required
# augmented dickey-fuller test removes autocorrelation and test for stationariry 
library(tseries)
x = rnorm(1000) # random normal data 
adf.test(x)
# here our p-value is below .05 which means stationarity, 
# means no further processing is nessasry for this model
```

```r
# test nottem set
adf.test(nottem)

#seasonality does not effect stationarity 
```

### autocorreclation in timeseries, describes the correctlation within the timeseries

    we wish to know if previous observations influence the later ones, corelation between lags
    
    autocorrecelation is often obvious
    
    in the lynx dataset there should be autocorrelation since if we trap many lynx one year there should be less lynx the next year
    
    

```r
# autocorrelation
plot(lynx)
```

```r
acf(lynx, lag.max=20) #autocorrelation function
# we can see that there are several bars ranging out of the
# .95 confidance interval this indicates autocorrelation within that lag 
# we can omit the plot if we use plot=0 argument
```

```r
pacf(lynx, lag.max=20) # partial autocorrelation function 
```

```r
# if we generate some random data we will not see any autocorreclation

acf(rnorm(500), lag.max=20)
# we can see from the graphs that there is not autocorrelation 
```

```r
pacf(rnorm(500), lag.max=20, plot=0)
```
